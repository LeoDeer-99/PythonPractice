Define functions used repeatedly

```python
def Col_Genre(df, col_i):
    col_genre = 0
    for j in range(df.shape[0]):
        if df.iloc[j,col_i] != 0:
            col_genre = df.iloc[j,i]
            break
    return col_genre

def Remove_Zeros(df):
    df = df.loc[:, (df != 0).any(axis=0)]
    df.columns = np.arange(0, df.shape[1], dtype=int)
    return df, df.shape[1] 
```

```python
df_size = 300
```

Import dataframe

It is badly formated

```python
df_movie = pd.read_csv("Pandas/moviesDB.csv")
df_movie.head()
```

Format the dataframe

First simple
Then do the Genres

```python
df_movie = df_movie.drop(columns=['Rating', 'movie'], axis=1)
df_movie = df_movie.rename(columns={'title': 'Title', 'genres': 'Genres', 'year': 'Year', 'RottenTomato': 'Rotten Tomato'})
df_movie = df_movie.sample(frac=1, random_state=1)
df_movie = df_movie.set_index(np.arange(0, df_movie.shape[0], dtype=int))
df_movie_full = df_movie.drop(columns=['Genres'], axis=1)

df_movie = df_movie.truncate(after=df_size-1)

df_movie.head()
```

We don't know how many unique genres we will have so we will use this as an upper bound

```python
buffer_cols = df_movie['Genres'].nunique()
```

We are then going to break up the Genres into a new table and sort that out

```python
df_genres = df_movie['Genres'].str.split('|', expand=True)
df_genres = df_genres.fillna(value=0)
```

Now the Genres are in a new table

```python
df_genres
```

We want to add extra columns for all the genres to have their own column when we sort them out. Then we will spread out the genres and into their own columns

```python
for i in np.arange(df_genres.shape[1], buffer_cols, dtype=int):    
    df_genres[i] = 0

col_size = df_genres.shape[1]

for i in range(col_size):
    for j in range(df_size):
        col_genre = Col_Genre(df_genres, i)
        if df_genres.iloc[j,i] != col_genre and df_genres.iloc[j,i] != 0:
            for l in range(col_size-i-1):
                df_genres.iloc[j,col_size-l-1] = df_genres.iloc[j,col_size-l-2]
            df_genres.iloc[j,i]=0
            
df_genres, col_size = Remove_Zeros(df_genres) 
```

```python
df_genres
```

We now want to cosolidate the duplicate rows into a single row. Then discretize the columns to 0 and 1 and label the column header by genres         

```python
# cosolidate the duplicate rows
for i in range(col_size):
    col_genre = Col_Genre(df_genres, i)
    for j in range(df_size):
        for i2 in range(col_size-i):
            if df_genres.iloc[j,i2+i] == col_genre and df_genres.iloc[j,i2+i] != 0:
                df_genres.iloc[j,i] = col_genre
                if i2 != 0:
                    df_genres.iloc[j,i2+i] = 0
                    
df_genres, col_size = Remove_Zeros(df_genres)     

# discretize the columns and label the column headers
for i in range(col_size):
    col_genre = Col_Genre(df_genres, i)
    df_genres = df_genres.rename(columns={i: col_genre})
    df_genres[col_genre] = df_genres[col_genre].replace(col_genre, 1)
```

```python
df_genres
```

Now merge this dataframe into the original dataframe and replace the genre tab.

```python
df_movie = pd.concat([df_movie.loc[:,'Title'], df_movie.loc[:,'Year':'Rotten Tomato'], df_genres], ignore_index=False, sort=False, axis=1)
```

## Plots

First we want to create some plots

* Pie and bar chart of genres breakdown
* Line graph of ratings through time

### Pie and bar chart of genres breakdown

```python
cmap = plt.get_cmap("tab20c")
df_genres.sum().sort_values(ascending=False).plot.pie(y='Genres', legend=True, title='Genres', labeldistance = 1.1, colors = cmap(np.arange(19)))

percents = df_genres.sum().sort_values(ascending=False) * 100 / df_genres.sum().sum()

plt.legend( bbox_to_anchor=(1.8,1.1), loc='upper right',
            labels=['%s, %1.1f %%' % (l, s) for l, s in zip(df_genres.sum().sort_values(ascending=False).index,percents)])

plt.show()
```

```python
percents = df_genres.sum().sort_values(ascending=False) * 100 / df_genres.sum().sum()

threshold = 3.
below_thresh_mask = percents < threshold
plot_data = percents[~below_thresh_mask]
plot_data.loc['Other'] = percents[below_thresh_mask].sum()
wedges, texts, autotexts = plt.pie(plot_data, labels=plot_data.index, autopct='%1.1f%%', pctdistance=0.85, colors = cmap(np.arange(11)))
plt.legend(bbox_to_anchor=(1.8,1.1), loc='upper right',
            labels=['%s, %1.1f %%' % (l, s) for l, s in zip(plot_data.index,plot_data)])
plt.title('Genres')
plt.show()
```

```python
df_genres.sum().sort_values(ascending=False).plot.bar(legend=False, figsize=(10,5))
plt.title('Genre Frequencies')
plt.xlabel('Genres')
plt.yticks(np.arange(0,141,10))
plt.xticks(rotation=90)
plt.show()
```

### Line graph of ratings through time

An additional graph that might be useful here would be the number of films per year this dataset shows.

```python
rating_by_year = df_movie_full.loc[:,['Year','Rotten Tomato']].groupby('Year').mean().reset_index()
films_per_year = df_movie_full.loc[:,['Year']].groupby(df_movie.loc[:,['Year']].columns.tolist(),as_index=False).size()

rating_by_year.sort_values(ascending=False, by='Year').plot(x="Year", y="Rotten Tomato", figsize=(10,4))
plt.xticks(np.arange(1900,2021,5), rotation=45)
plt.xlabel('Year')
plt.yticks(np.arange(50,100,5))
plt.ylabel('Rotten Tomato Score')
plt.title('Rotten Tomato Score Over Time')

films_per_year.sort_values(ascending=True, by='Year').plot(x="Year", y="size", figsize=(10,4), legend=False)
plt.xticks(np.arange(1900,2021,5), rotation=45)
plt.xlabel('Year')
plt.yticks(np.arange(0,320,20))
plt.ylabel('Frequency')
plt.title('Films per Year')
plt.show()
```
